# Выполнение

```python
reviews = ["""Телефон абонента: +7 916 000 00 01
Время: 10:34
Дата: 20.05.2017
Оценка: 1
Длительность разговора (мин.): 0:32
Комментарий: Общение с сотрудником вашей техподдержки мне не понравилось. Раньше всегда отвечала девушка с приятным голосом, а сегодня какой-то мужчина. Совсем никуда не годится, примите меры!!!""",
"""Телефон абонента: +7 916 000 00 02
Время: 9:20
Дата: 21.05.2017
Оценка: 5
Длительность разговора (мин.): 25:07
Комментарий: С моей проблемой мне помогли, большое спасибо!"""]
```

### Первый вариант решения, тут мы заявзаны на пробел после ": "

```python
my_base = dict() # создаем пустой словарь my_base

for item in reviews: # перебираем каждый элемент в списке rewiews. 

    dictionary_key = item.split("\n")[0].split(": ")[1] 
    # выдергиваем номер телефона и присваиваем его переменной dictionary_key
    # если разбирать подробней, то берем первую строку в каждом элементе списка и берем первую запись в этой строке через split
    
    my_base[dictionary_key] = dict() 
    # создаем элемент словаря my_base состоящий из пустого подсловаря с именем = номеру телефона
    

    for line in item.split("\n"): # для каждой строки в каждом элементе списка reviews
        name_value = line.split(": ")[0] # присваиваем первый элемент как заголовок
        data_value = line.split(": ")[1] # присваиваем второй элемент как значение
        my_base[dictionary_key][name_value] = data_value # заполняем наш словарь проходя цикл

        

print()
print()
print(my_base)
print("Проверка условия задания:", (my_base["+7 916 000 00 01"]["Дата"] == "20.05.2017"))
```

### Альтернативное решение, не зависящее от пробела после :

```python
my_base = dict() # создаем пустой словарь my_base

for item in reviews: # перебираем каждый элемент в списке rewiews. 

    dictionary_key = item.split("\n")[0].split(":")[1].strip()
    # выдергиваем номер телефона и присваиваем его переменной dictionary_key
    # методом strip() удаляем лишний пробел перед номером телефона, так как разбивали просто по ":" а не по ": "
    # если разбирать подробней, то берем первую строку в каждом элементе списка и берем первую запись в этой строке через split
    
    my_base[dictionary_key] = dict() 
    # создаем элемент словаря my_base состоящий из пустого подсловаря с именем = номеру телефона
    

    for line in item.split("\n"): # для каждой строки в каждом элементе списка reviews
        name_value = line.split(":")[0] # присваиваем первый элемент как ключ словаря
        data_value = ":".join(line.split(":")[1:]).strip()
        # присваиваем второй элемент как значение предварительно склеив его в одну сущность
        # подробнее: сначала мы разбиваем строку на список по ":", 
        # и джойним его обратно в строку выкидывая первый элемент и убирая лишние пробелы
        
        my_base[dictionary_key][name_value] = data_value # заполняем наш словарь проходя цикл

        

print()
print()
print(my_base)
print("Проверка условия задания:", (my_base["+7 916 000 00 01"]["Дата"] == "20.05.2017"))
```

### Третий вариант решения (преподавателем)

```python
my_bose = {} # Создаем словарь присваивая переменной пустое значение

for review in reviews: # перебираем каждый элемент списка reviews
    
    review_dict = {} # создаем словарь чтобы потом вложить его в наш основной словарь и на каждом шаге цикла обнуляем его
  
    for line in review.split("\n"): # перебираем каждый элемиент списка reviews деля его построчно
        tokens = line.split(':') 
        # каждую строку сплитим по ":" и создаем список tokens 
        # при делении по : будет много мусорных элементов списка, это надо будет починить
     
        name = tokens[0]
        # первый элемент списка tokens это заголовок элемента для нашего словаря, присваиваем его переменной name
        
        remained_part = ':'.join(tokens[1:]).strip()
        # список мог некорректно поделиться так как в тексте приссутствовали мусорные ":"
        # для этого мы джойним список tokens обратно в строку, начиная со второго элемента меняя все разделители на ":"
        # метод .strip удаляет пустые начальные пробелы и конечные пробелы
                
        review_dict[name] = remained_part 
        # в словаре review_dict создаем запись на каждой итерации цикла в связке ключ-значение
        
    my_bose[review_dict["Телефон абонента"]] = review_dict
    # на этом шаге мы задаем ключ словаря my_bose как значение review_dict["Телефон абонента"] в текущей итерации цикла
    # и присваиваем этому ключу целый словарь review_dict целиком как есть в этой итерации цикла
    # при заходе на следующую итерацию словарь review_dict обнулится


print(my_bose)
print("Проверка условия задания:", (my_bose["+7 916 000 00 01"]["Дата"] == "20.05.2017"))
```
