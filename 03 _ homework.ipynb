# Выполнение

```python
reviews = ["""Телефон абонента: +7 916 000 00 01
Время: 10:34
Дата: 20.05.2017
Оценка: 1
Длительность разговора (мин.): 0:32
Комментарий: Общение с сотрудником вашей техподдержки мне не понравилось. Раньше всегда отвечала девушка с приятным голосом, а сегодня какой-то мужчина. Совсем никуда не годится, примите меры!!!""",
"""Телефон абонента: +7 916 000 00 02
Время: 9:20
Дата: 21.05.2017
Оценка: 5
Длительность разговора (мин.): 25:07
Комментарий: С моей проблемой мне помогли, большое спасибо!"""]
print(type(reviews))
print(reviews)
```

```python
my_base = dict()
for item in reviews:

    dictionary_key = item.split("\n")[0].split(": ")[1]
    my_base[dictionary_key] = dict()

    for line in item.split("\n"):
        name_value = line.split(": ")[0]
        data_value = line.split(": ")[1]
        my_base[dictionary_key][name_value] = data_value


print("Созданный словарик:")
print(my_base)
print()
print("Проверка условия задания:", (my_base["+7 916 000 00 01"]["Дата"] == "20.05.2017"))
```

```python
my_bose = {} # Создаем словарб присваивая переменной пустое значение
for review in reviews:
    review_dict = {} #  создаем словарь для новой записи
    for line in review.split("\n"):
        # мы видим, что имена полей отделяются двоеточием, поэтому тоже сделаем split по этому символу
        tokens = line.split(':') 
        name = tokens[0] # первая часть будет названием поля
        # в оставшейся части могли быть символы двоеточия, а мы по ним разделили, поэтому нужно обратно
        remained_part = ':'.join(tokens[1:]).strip() # метод .strip удаляет начальные пробелы и конечные пробелы
        # в словаре review_dict создаем нужную запись
        review_dict[name] = remained_part
        # в итоговом словаре ключом является телефон абонента, поэтому добавим запись по этому ключу
    my_bose[review_dict["Телефон абонента"]] = review_dict

print("Созданный словарик:")
print(my_bose)
print()
print("Проверка условия задания:", (my_bose["+7 916 000 00 01"]["Дата"] == "20.05.2017"))
```
