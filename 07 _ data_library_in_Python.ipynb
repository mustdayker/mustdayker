```python
df.loc[[1, 3, 4], 'YrSold'] # в данном примере мы получили 1,3 и 4 значение в колонке YrSold
```

# Расчет произвольных функций

## Для Series

```python
def price_class(price): # Создаем функцию price_class которая возвращает значение в зависимости от условия
    return "Низкий" if price < 200000 else "Высокий" # Если значение меньше 200000 и  в ином случае

df.SalePrice.apply(price_class).head(5) # Применяем функцию price_class к колонке SalePrice
```

## Для DataFrame

```python
def interesting(row): # Создаем функцию interesting, которая определяет интесно нам или нет
    if (row['PoolArea'] > 0) and (row['SalePrice'] < 250000): # Если есть басик и прайс < 250000
        return "Интересна!!! Есть басик!!" # Возвращаем текстовое значение, что нам интересно
    elif row['SalePrice'] < 200000: # Если прайс ниже чем 200000 
        return "Интересна, но без басика." 
    else:
        return "Прайс - борщ!!" # Иначе возвращаем что прайс слишком большой

df.apply(interesting, axis=1).head(5) # Применяем функцию к вертикальной оси таблицы
                                        # Так как в функции описано какие колонки обходить, то она пойдет по ним
```

```python
df.filter(like='Garage') # выводит только те заголовки в которых есть сочетание символов "Garage"
```

```python
df.filter(regex='Yr|Year') # Можно фильтровать передавая сочетания значений, разделяя их |
```

## Фильтрация по содержимому
Используем индексацию и булевы операции

```python
df.SaleCondition == 'Normal'
```

```python
df.loc[df.SaleCondition == 'Normal'] # df.SaleCondition == 'Normal' вернет нам Series из True и False
                                    # После чего функция loc профильтрует нашу таблицу и вернет только те
                                    # строки где SaleCondition == Normal (то есть вернулось True)
```

```python
df.loc[df.SalePrice < 150000, 'Fence'] # Выводит ячейки колонки Fence для которых в колонке SalePrice выполняется условие
                                        # SalePrice < 150000
                                        # Результат это Series
```

```python
df.loc[df.SalePrice < 150000, ['PoolArea', 'Fence', 'SalePrice']] # Выводит ячейки колонки 'PoolArea', 'Fence', 'SalePrice' 
                    # для которых в колонке SalePrice выполняется условие < 150000
                    # Результат это DataFrame
```

## Сортировка по индексам

```python
df.sort_index() # Сортировать таблицу по индексу в алфавитном порядке
```

```python
df.sort_index(ascending=False) # Сортировать таблицу по индексу в обратном порядке
```

```python
df.sort_index(axis=1) # Сортировать таблицу по названию столбцов в алфавитном порядке
```

## Сортировка по значениям

```python
df.sort_values('SalePrice') # Сортировать по значениям колонки SailPrice по возрастанию
```

```python
df.sort_values(['SaleCondition', 'SalePrice'])[['SaleCondition', 'SalePrice']].to_csv('Train_Sorted.csv')
                            # Сортировать по значениям двух колонок, сортировка происходит по очереди
                            # И записать результат работы этого файла в csv файл
```
