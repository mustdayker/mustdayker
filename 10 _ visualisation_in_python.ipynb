## Много графиков на одной плоскости

### Сгенерируем данные

```python
# Сгенерируем данные
from math import log
```

```python
linear = list(range(1, 11))
quadratic = [x**2 for x in range(1, 11)]
logarithmic = list(map(log, range(1, 11)))
```

```python
print('lineaar:', linear)
print('quadratic:', quadratic)
print('logarithmic:', logarithmic)
```

## Построим график

```python
plt.plot(linear, 'b-')
plt.plot(logarithmic, 'g-')
plt.plot(quadratic, 'r-')
```

```python
plt.plot(linear, 'b-')
plt.plot(logarithmic, 'g-')
plt.plot(quadratic, 'r-')
plt.legend(['linear', 'quadratic', 'logarithmic'], loc = 2) # НЕ ПРАВИЛЬНО, так как легенда передается в порядке написания
```

```python
plt.plot(linear, 'b-', label = 'linear')
plt.plot(logarithmic, 'g-', label = 'logarithmic')
plt.plot(quadratic, 'r-', label = 'quadratic')
plt.legend(loc = 2)
```

```python
plt.plot(linear, 'b-', linewidth = 2.0, label = 'linear')
plt.plot(logarithmic, 'g-', linewidth = 2.0, label = 'logarithmic')
plt.fill_between(range(10), logarithmic, linear, color = 'blue', alpha = 0.2, label = 'place between')
plt.legend(loc = 2)
```

# Точки, гистограммы, диаграммы, распределения

```python
import pandas as pd
from sklearn import datasets # этот модуль содержит наборы данных для обучения
```

```python
iris = datasets.load_iris() # загружаем датасет IRIS
```

```python
print(iris) # смотрим что внутри датасета
```

```python
print(iris.DESCR) # описание датасета
```

```python
iris.data # Чтобы получить только данные с датасета надо вызвать .data с этого датасета (в нашем случае из переменной)
```

```python
iris.target_names # названия трех классов цветков
```

```python
iris.target # метка, которая говорит о том что каждый объект имеет свой собственный тип
```

```python
iris.feature_names
```

```python
iris_frame = pd.DataFrame(iris.data) # Собираем DataFrame из iris.data
iris_frame.columns = iris.feature_names # В названия колонок передает iris.feature_names
iris_frame['target'] = iris.target # Добавим столбец target куда пропишем классификаторы
iris_frame.target = iris_frame.target
```

```python
iris_frame.head(10)
```

```python
iris_frame.describe() # краткая статистика по датафрейму
```

# Диаграммы

```python
iris_frame.target.value_counts() # посмотрим сколько объектов каждого типа у нас есть
```

```python
plt.figure(figsize=(5,5))
plt.pie(iris_frame.target.value_counts(), labels=iris['target_names'])
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(iris_frame.target.value_counts(), autopct='%1.1f%%', labels=iris['target_names'])
                                # % %% - говорит от том что у нас будет форматированный вывод
                                # 1.1f - указываем число с точностью 1 символ после запятой f - в формате float
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(iris_frame.target.value_counts(), explode=[0, 0.2, 0], autopct='%1.1f%%', 
        shadow=True, startangle=90, labels=iris['target_names'])
plt.show()
```

# Гистограммы

```python
iris_frame[iris_frame.target == 0].hist('sepal length (cm)')
```

```python
plt.hist(iris_frame.loc[iris_frame.target == 0, 'sepal length (cm)'], bins=10)
                    # bins количество столбиков
```

```python
plt.figure(figsize=(16,4))

plt.subplot(131)
plt.hist(iris_frame.loc[iris_frame.target == 0, 'sepal length (cm)'])
plt.title('setosa')

plt.subplot(132)
plt.hist(iris_frame.loc[iris_frame.target == 1, 'sepal length (cm)'])
plt.title('versicolor')

plt.subplot(133)
plt.hist(iris_frame.loc[iris_frame.target == 2, 'sepal length (cm)'])
plt.title('virginica')
```

```python
plt.figure(figsize=(10,6))
plt.title('Sepal length (cm)')
plt.hist(iris_frame.loc[iris_frame.target == 0, 'sepal length (cm)'], alpha = 0.5, label='setosa')
plt.hist(iris_frame.loc[iris_frame.target == 1, 'sepal length (cm)'], alpha = 0.5, label='versicolor')
plt.hist(iris_frame.loc[iris_frame.target == 2, 'sepal length (cm)'], alpha = 0.5, label='vieginica')
plt.legend(loc='upper right')
```

```python
plt.figure(figsize = (20, 24))

plot_number = 0
for feature_name in iris['feature_names']:
    for target in iris_frame.target.unique():
        plt.subplot(4, 3, plot_number + 1)
        plt.hist(iris_frame[iris_frame.target == target][feature_name])
        plt.title(iris['target_names'][target])
        plt.xlabel('cm')
        plt.ylabel(feature_name[:-4])
        plot_number += 1
```
